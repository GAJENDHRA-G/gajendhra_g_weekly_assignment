## Q1. What is the difference between `is` operator and `==` operator?

### Answer

The `==` operator checks whether two variables have the same value, whereas the `is` operator checks whether both variables refer to the same object in memory.

**Example**

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)
print(a is b)
```

Output

```
True
False
```

## Q5. Why is dictionary lookup O(1) while list search is O(n)?

### Answer

A dictionary directly finds the required key using hashing, so it usually takes constant time.

A list checks each element one by one until it finds the required value, so the time increases with the size of the list.

````python
# Q27 - Number Diamond Pattern

n = int(input("Enter N : "))

# Upper Half
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()

# Lower Half
for i in range(n - 1, 0, -1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()
````

````python
# Q31 - Armstrong Numbers between 1 and 9999

count = 0

print("Armstrong Numbers:")
for num in range(1, 10000):

    temp = num
    digits = len(str(num))
    total = 0
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10
    if total == num:
        print(num, end=" ")
        count += 1
print("\n")
print("Total Armstrong Numbers =", count)
````

````python
# Q32 - Fibonacci Series

n = int(input("Enter Number of Terms : "))

first = 0
second = 1
total = 0

if n == 0:
    print("Sum = 0")
elif n == 1:
    print(first)
    print("Sum =", first)
else:
    print("Fibonacci Series:")
    for i in range(n):
        print(first, end=" ")
        total += first
        next_num = first + second
        first = second
        second = next_num
    print("\nSum =", total)
````

````python
# Q33 - Prime Number and Sum of Prime Numbers

n = int(input("Enter a Number : "))
is_prime = True

if n < 2:
    is_prime = False
else:
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

if is_prime:
    print(n, "is a Prime Number")
else:
    print(n, "is not a Prime Number")

prime_sum = 0

for num in range(2, n + 1):
    prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break
    if prime:
        prime_sum += num

print("Sum of Prime Numbers =", prime_sum)
````

````python
# Q34 - Reverse Number and Palindrome

number = int(input("Enter Number : "))

if number < 0:
    print("Negative numbers are not palindrome.")
else:
    original = number
    reverse = 0
    while number > 0:
        digit = number % 10
        reverse = reverse * 10 + digit
        number //= 10
    print("Reversed Number =", reverse)

    if original == reverse:
        print("Palindrome")
    else:
        print("Not Palindrome")
````

````python
# Q35 - HCF and LCM

a = int(input("Enter First Number : "))
b = int(input("Enter Second Number : "))

x = a
y = b

while y != 0:
    remainder = x % y
    x = y
    y = remainder

hcf = x
lcm = (a * b) // hcf

print("HCF =", hcf)
print("LCM =", lcm)
````

````python
# Q36 - String Scanner

text = input("Enter a Sentence : ")

characters = 0
words = len(text.split())
vowels = 0
consonants = 0
digits = 0
spaces = 0
special = 0

for ch in text:

    characters += 1

    if ch.lower() in "aeiou":
        vowels += 1

    elif ch.isalpha():
        consonants += 1

    elif ch.isdigit():
        digits += 1

    elif ch.isspace():
        spaces += 1

    else:
        special += 1

print("\nCharacters :", characters)
print("Words :", words)
print("Vowels :", vowels)
print("Consonants :", consonants)
print("Digits :", digits)
print("Spaces :", spaces)
print("Special Characters :", special)
````

````python
# Q41 - TCS Stock Price Analysis

prices = (3845, 3902, 3878, 3955, 4012, 3998, 4055)
days = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")

maximum = max(prices)
minimum = min(prices)

print("Maximum Price :", maximum)
print("Day :", days[prices.index(maximum)])

print()

print("Minimum Price :", minimum)
print("Day :", days[prices.index(minimum)])

average = sum(prices) / len(prices)

print()

print("Average Price :", round(average, 2))

gain = 0
gain_day = ""

for i in range(1, len(prices)):
    difference = prices[i] - prices[i - 1]
    if difference > gain:
        gain = difference
        gain_day = days[i - 1] + " to " + days[i]
print()

print("Biggest Gain :", gain)
print("Between :", gain_day)

volatility = 0

for i in range(1, len(prices)):
    volatility += abs(prices[i] - prices[i - 1])
print("Total Volatility :", volatility)
````

````python
# Q45 - Caesar Cipher

message = input("Enter Message : ")
shift = int(input("Enter Shift Value : "))
encrypted = ""

for ch in message:
    if ch.isalpha():
        if ch.isupper():
            encrypted += chr((ord(ch) - 65 + shift) % 26 + 65)
        else:
            encrypted += chr((ord(ch) - 97 + shift) % 26 + 97)
    else:
        encrypted += ch

print()
print("Encrypted Message")
print(encrypted)
````

````python
# Q46 - GST Calculator

def calculate_gst(amount, gst_rate=18, gst_inclusive=False):
    if gst_inclusive:
        base = amount / (1 + gst_rate / 100)
        gst = amount - base
        total = amount
    else:
        base = amount
        gst = amount * gst_rate / 100
        total = amount + gst
    return round(base,2), round(gst,2), round(total,2)


print(calculate_gst(1000))
print(calculate_gst(1180, gst_inclusive=True))
print(calculate_gst(500, gst_rate=5))
````

````python
# Q47 - Recursive Power Function

def power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent < 0:
        return 1 / power(base, -exponent)
    else:
        return base * power(base, exponent - 1)

print(power(2,10))
print(power(5,0))
print(power(3,-2))
````
